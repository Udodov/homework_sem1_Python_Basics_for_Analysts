### Задание 1. Найдите, какая выручка у компании в ноябре.  Используйте list comprehensions.

Объявим исходные списки

````python
date_purchase = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09',]

amount = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887,]
````

Затем, используя *list comprehensions* создадим список, который пробгается по этому массиву, выбирая суммупо условию выборку для ноября: разбивая с первого элемента на части по разделителю "-" и выбрав вторую часть (месяц), которую проверяем на равность 11 ```month[0].split('-')[1]```. ### Задание 2. Найдите выручку компании в зависимости от месяца. Напишите функцию, которая на вход принимает списки с датами и с выручкой, а на выходе вернет словарь, где ключи - это месяцы, а значения - это выручка.

Создадим функцию, которая вычисляет выручку по месяцам на основе двух списков: **date_purchase** и **amount**. 

Первый цикл for проходит по всем элементам списка **date_purchase**. Для каждого элемента извлекается месяц, т.е. делаем срез (подстрока с 5-го по 6-й символы) и проверяется, есть ли он уже в словаре **dict1**. Если месяца еще нет в словаре, то создается новая запись с ключом равным месяцу и значением равным соответствующему элементу списка **amount**. Если же месяц уже есть в словаре, то к соответствующему значению добавляется значение из списка **amount**.

В конце функция возвращает словарь **dict1**, в котором ключами являются месяцы, а значениями - выручка за каждый месяц.

````python
def revenue(date_purchase: list, amount: list) -> dict:
    dict1 = {}

    for i in range(len(date_purchase)):
        month = date_purchase[i][5:7]

        if month not in dict1:
            dict1[month] = amount[i]

        else:
            dict1[month] += amount[i]

    return dict1


revenue(date_purchase, amount)
````
